As we managed to deal with technical difficulties we could proceed with the analysis. Firstly we write a method to get the relative price increment (or return) $\eta_{k}$ from a price serie following formula $(6.4)$. (We implement it for both pandas series and lists)

```python
def price_to_returns(series):
    if (type(series)==list):
        series_1=series[1:]
        series_2=series[:-1]
    else:
        series_1=series.iloc[1:].values
        series_2=series.iloc[:-1].values
    return np.log(series_2)-np.log(series_1)
    
```

next we implement a method calculate the covariance. By adding elements on the diagonal of the covariance matrix we get variance, adding values in the $l$th skew axis above the diagonal we get $\langle\eta_{k}\eta_{k+l}\rangle$

```python
def covariance_calculation(returns,k):
    cov=0
    j=0
    for i in range(0,len(returns)-k):
        cov+=returns[i]*returns[i+k]
        j+=1
    cov=cov/j
    return cov
```

And finaly we write a full method to calculating correlations $l$ stands for the largest $C(l)$ argument i.e. time slice $t=l\cdot\tau=l\cdot 5min$

```python
 def corelations_1(l):
    C_l=[0]*l
    N=0
    for year in range(1,3):
        for month in range(1,13):
            slice_name='year'+str(year)+'month'+str(month)
            df = pd.read_csv('extended_intraday_SPY_5min_'+slice_name+'_raw.csv')
            series=df.iloc[:,4]
            returns=price_to_returns(series)
            N+=len(returns)
            C_l_month=[]
            for tau in range(0,l):
                cov=covariance_calculation(returns, tau)
                C_l_month.append(cov)
            C_l=[sum(x) for x in zip(C_l, C_l_month)]
    #we are missing contributions to correlations where etas are form different months I think it won't change the picture
    #still we have overnight 
    #normalisation
    C_l = [item /24 for item in C_l]
    for i in range(1,l):
        C_l[i]=C_l[i]/C_l[0]
    return C_l[0],C_l[1:],N
```

Now we calculate variogram of logarithms of prices $V(l)$ from equation $(6.6)$ (here the method only works for previously downloaded data)

```python
def log_price_variograms(log_price,l):
    v=0
    j=0
    for i in range(0,len(log_price)-l):
        v+=(log_price[i]-log_price[i+l])**2
        j+=1
    v=v/j
    return v
```

```python
 def variogram():
    V_l=[0]*100
    for year in range(1,3):
        for month in range(1,13):
            slice_name='year'+str(year)+'month'+str(month)
            df = pd.read_csv('extended_intraday_SPY_5min_'+slice_name+'_raw.csv')
            series=df.iloc[:,4]
            log_price=np.log(series.values)
            V_l_month=[]
            l=[]
            l.append(1)
            while l[len(l)-1]<1000:
                var=log_price_variograms(log_price,l[len(l)-1])
                l.append(l[len(l)-1]+10)
                V_l_month.append(var)
            V_l=[sum(x) for x in zip(V_l, V_l_month)]
            l=l[:-1] 
    #we are missing contributions to correlations where data are form different months I think it won't change the picture
    #still we have overnight 
    #normalisation
    V_l = [item /24 for item in V_l]
    return V_l ,l
```

Now we calculate cumulative probability distributions for uncorrelated $\eta$s (therefore we choose $\tau=30min$). For positive $\eta$s we are interested in $P(H>\eta)$ whereas for negative $\eta$s in $P(H<\eta)$. By capital $H$ we mean here the random variable and by small one the particular realisation. 

```python
def cumulative_distribution():
    price=[]
    for year in range(1,3):
        for month in range(1,13):
            slice_name='year'+str(year)+'month'+str(month)
            df = pd.read_csv('extended_intraday_SPY_5min_'+slice_name+'_raw.csv')
            price_5=df.iloc[:,4].tolist()
            price_30=price_5[0::6] #we take \tau larger that correlation time here we take \tau=30min 
            price+=price_30
    returns=price_to_returns(price)
    positive=[]
    negative=[]
    for i in returns:
        if i>=0:
            positive.append(i)
        else:
            negative.append(i)
    positive.sort()
    negative.sort(reverse=True)
    prob_pos=[]
    val=0.0
    num=len(positive)
    j=0
    for i in range(0,len(positive)):
        if(positive[i]>val):
            val=positive[i]
            num=num-j
            j=0
        j+=1
        prob_pos.append(num)
    prob_neg=[]
    val=0.0
    num=len(negative)
    j=0
    for i in range(0,len(negative)):
        if(negative[i]<val):
            val=negative[i]
            num=num-j
            j=0
        j+=1
        prob_neg.append(num)
    prob_pos = [item /prob_pos[0] for item in prob_pos]
    prob_neg = [item /prob_neg[0] for item in prob_neg]
    var_pos=np.var(list(set(prob_pos)))
    var_neg=np.var(list(set(prob_neg)))
    return positive, prob_pos, negative, prob_neg,var_pos,var_neg #powtarzające się wartości mocno zawyżają wariację poprawić 
```

```python
positive, prob_pos, negative, prob_neg,var_pos,var_neg=cumulative_distribution()
```

Now we will try to fit different functions to that cumulative distribution as was done in the book. We start with Gaussian. 

```python
def gauss(x,mu,sigma,A):
    return A*np.exp(-(x-mu)**2/2/sigma**2)
```

```python
from scipy.optimize import curve_fit
params,cov=curve_fit(gauss,positive,prob_pos,bounds=((-5,0.0,0.5), (5,0.005,500000)),maxfev=50000)
#params,cov=curve_fit(lambda x, mu, A: gauss(x,mu,np.sqrt(var_pos),A),positive,prob_pos,bounds=((-5,0.5), (5,1.2)))
print(params)
```

And then Student distribution 

```python
from scipy.stats import t
params_t,cov_t=curve_fit(lambda x, df,loc,scale: t.pdf(x,df,loc,scale),positive,prob_pos,maxfev=50000)
pdf_fitted = t.pdf(positive, loc=params_t[1], scale=params_t[2], df=params_t[0])
mean, var, skew, kurt = t.stats(loc=params_t[1], scale=params_t[2], df=params_t[0], moments='mvsk')
print(params_t)
print(mean, var, skew, kurt)
```
